### MINI-SUPERMARKET WEEKLY ANALYSIS

This notebook focuses on the weekly analysis of Fake Random Data generated by me for A small shop that sells confectionaries. Number of items sold daily for each product ranged from 15 to 49 for a period of seven days. `The total income for the shop  per day` will be calculated as well as the `total income for the week`. `Average nummber of each product sold per week` will be calculated as well.


#### DATASET

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()


# Setting random seed for consistency
np.random.seed(42)

product_names = ["Bread", "Milk", "Soda", "Canned Water", "Lolipop", "Peanut", "Groundnuts", "Wine"]
days_of_the_week = ["Sun", 'Mon', 'Tues', "Wed", 'Thurs', "Fri", "Sat"]
food_prices = pd.DataFrame(np.array([10, 25, 15, 10, 20, 10, 20, 40]).reshape(1,8), columns=product_names, index=["Prices"])
food_sold = pd.DataFrame(np.random.randint(15, 49, size=(7,8)), columns=product_names, index=[days_of_the_week])

print(food_prices)
print(food_df)
```

####  EXPLORATORY DATA ANALYSIS ON DAILY SALES

```python
# Total amount sold in a day for all the food items
daily_sales = np.dot(food_sold, food_prices.T)
food_sold["Daily_sales"] = daily_sales

food_sold
```

```python
# Avearge product sold per day
bread_sold_average = np.mean(food_sold["Bread"])
milk_sold_average =  np.mean(food_sold["Milk"])
soda_sold_average = np.mean(food_sold["Soda"])
canned_water_sold_average = np.mean(food_sold["Canned Water"])
lolipop_sold_average = np.mean(food_sold["Lolipop"])
peanuts_sold_average = np.mean(food_sold["Peanut"])
groundnut_sold_average = np.mean(food_sold["Groundnuts"])
wine_sold_average = np.mean(food_sold["Wine"])



average = [bread_sold_average, 
           milk_sold_average, 
           soda_sold_average, 
           canned_water_sold_average,
           lolipop_sold_average, 
           peanuts_sold_average, 
           groundnut_sold_average, 
           wine_sold_average]

average_df = pd.DataFrame({"Products": product_names, "Average-in-the-week":average})

fig, ax = plt.subplots()
ax = ax.barh(average_df["Products"], average_df["Average-in-the-week"], color=["r", "b", "m", "k", "y", "g", "c"])
```

Interpretation from the graph:
* On average, Bread and peanut sold the highest daily, followed closely by Milk and Groundnuts. Funny enough soda sold the lowest that week below wine.

This can be viewed in a `Pie Chart` below

Though the bar chart shows the most undertandable results, the pie chart below was still produced similar results but not one that could be understood easily, it is even better suited for aesthetic purposes
